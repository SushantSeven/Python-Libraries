5. Write a Python program to plot two or more lines on same plot with suitable legends of each
line.

```python
age_men = [30, 35, 40, 45, 50, 55, 60, 65, 70]
blood_sugar_men = [45, 66, 82, 90, 105, 111, 115, 120, 125]

age_women = [30, 35, 40, 45, 50, 55, 60, 65, 70]
blood_sugar_women = [80, 85, 89, 95, 100, 105, 110, 111, 119]

plt.plot(age_men,blood_sugar_men, label = 'men')
plt.plot(age_women,blood_sugar_women, label = 'women')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Blood sugar level')
plt.title('Blood sugar level for men and women')
plt.show()
```

15. Write a Python program to create multiple plots.

```python
X = np.arange(0, math.pi*2, 0.05) 
Y1 = np.sin(X) 
Y2 = np.cos(X) 
Y3 = np.tan(X) 
Y4 = np.tanh(X)

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2, sharex=True)
fig.set_size_inches(10.5, 7.5)

ax1.plot(X,Y1)
ax1.set_title('sin')
ax1.set_ylabel('y')

ax2.plot(X,Y2, color='r')
ax2.set_ylabel('y')
ax2.set_title('Cos')

ax3.plot(X,Y3, color='g')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_title('Tan')

ax4.plot(X,Y4, color='black')
ax4.set_xlabel('x')
ax4.set_ylabel('y')
ax4.set_title('Tanh')
plt.show()
```
