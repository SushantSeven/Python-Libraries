1. Write a program to draw a scatter plot for random 1000 x and y coordinates

```python
import numpy as np
import plotly.express as px
import pandas as pd

x_coordinates = np.random.randint(0,100,size=1000)
y_coordinates = np.random.randint(0,100,size=1000)

px.scatter(x=x_coordinates, y=y_coordinates)
```

2. Write a program to draw line and scatter plots for random 100 x and y coordinates

```python
x1_coordinates = np.random.randint(0,100,size=100)
y1_coordinates = np.random.randint(0,100,size=100)
px.line(x=x1_coordinates, y=y1_coordinates)
```

3. Write a program to draw a scatter plot for random 500 x and y coordinates and style it

```python
x3_coordinates = np.random.randint(0,100,size=500)
y3_coordinates = np.random.randint(0,100,size=500)

px.scatter(x=x3_coordinates, y=y3_coordinates, color=x3_coordinates,size=x3_coordinates)    
```
