1. Write a program to draw a scatter plot of “day” against “total bill” for a dataset given in a
url

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

total_bill_df = pd.read_csv('total_bil.csv')
sns.scatterplot(data=total_bill_df, x='day',y = 'total_bill')
plt.show()
```
