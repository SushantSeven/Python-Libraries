3. Write a Python program to add, subtract, multiple and divide two Pandas Series.
Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 9]

```python
q3series1 = pd.Series([2, 4, 6, 8, 10])
q3series2 = pd.Series([1, 3, 5, 7, 9])

print("The original series",q3series1.values,q3series2.values)

print("Addition of two series",(q3series1.add(q3series2)).values)
print("Subtraction of two series",(q3series1.sub(q3series2)).values)
print("Mutiplication of two series",(q3series1.mul(q3series2)).values)
print("Division of two series",(q3series1.div(q3series2)).values)
```

4. Write a Python program to get the powers of an array values element-wise.
Note: First array elements raised to powers from second array
Expected Output:
Original array
[0 1 2 3 4 5 6]
